# Below we will be implementing QQ plot which is a way to graphically determine whether a data is normally distributed or not

```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(10,6)) #defining the overall plot size.
    plt.subplot(1,2,1) #defining first subplot
    sns.histplot(sample,kde=True) #Using histogram to check normality of distribution in first subplot
    plt.subplot(1,2,2)  #defining second subplot
    stat.probplot(sample,dist='norm',plot=pylab) #Using qq plot to check normality of distribution in second subplot
    plt.show() #displaying whole plot created
```

```python
#1. In histogram: kde function can be used to judge normal distribution.
#2. In qq plot if all the data points lie on red line(which is hypothesis line for data to be normally distributed) then data
#   is normally distributed.In this way qq plot can be used to judge normal distribution

#Case 1: Using histogram and qq plot to check normality of for NORMALLY DISTRIBUTED DATA
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
#From below fig we can conclude that variable s is normally distributed.
```

```python
#Case 2: Using histogram and qq plot to check normality for LOG NORMALLY DISTRIBUTED DATA
p=np.random.lognormal(3,1,1000)
plot_data(p)
#From below fig we can conclude that variable p is not normally distributed.
```

```python
plot_data(np.log(p))
#From below figure we can conclude that when we take natural logarithm of log normally distributed data then it get's converted
#into normally distributed data.
```

```python
plot_data(np.log(p))
```
