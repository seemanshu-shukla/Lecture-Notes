# 3. DATA AUGUMENTATION:
 - Here we are using the same CNN architecture and dataset then we used in Dog_Vs_Cat.ipynb  <br><br>
 
 -  In CNN for reducing the overfitting(Training Accuracy higer than Testing Accuracy) we can use DATA AUGUMENTATION. Other ways to counter overfitting can be Hyperparameter tuning, Dropout, Regularization  <br><br>
 
 - Jeffry Hinton introduced the concept of Data Augmentation in 2012 when then created the Alexnet architecture in the ImageNet competition <br><br>
 
 - Understanding Data Augumentation with the help of real life example:
   - During training CNN model to recognize Cat, we feeded model with the image of Cats standing in 4 legs whereas, during testing we passed the images Cat is standing in 2 legs or only 2 legs are visible or sleeping cat where no leg is visible. Then our CNN model can be prone to overfiting and can miss classify the different variations of Cat images that are present in the testing dataset. This is happening since Model only learned the pattern where cat is standing on 4 legs and not on 2 legs etc. 
   - So here Data Augumentation comes into rescue where we are basically, trying to Augument the training image of cat standing on 4 legs into new images of same cat but capturing the different variations like generating the image of cat standing in 2 legs, cat standing upside down etc so that model can predict the same when feeded with the same during testing
   - This is what we call as Data Augumentation that is generating the new augumented data which is synthetic data capured from different angles of the original data and using them as the training dataset so that model can capture the different patterns and variantions so that testing accuracy can be increased and model is less prone to the overfitting.
   - Please note that Data Augumentation is basically used in the cases where we have limited data in the training dataset
   - Also Data Augumentation is applied only on top of Training Dataset and not on the Testing Dataset

### In below cell we have prepared ImageGenerator object for following:
     1. ImageGenerator object for TRAINING DATASET: Here we are defining different attributes based on which DATA AUGUMENTATION will happen
     
     2. ImageGenerator object fot TESTING DATASET: Please note that Data Augumentation is only done on top of train dataset so that our model can capture different miscellaneous variations which during training, model is not aware of but while TESTING using TESTING DATASET we are providing these datapoints with different variantions to the model which can lead to poor test accuracy (or overfitting). Therefore, while using ImageGenerator object on top of TESTING DATASET we will not use any attribute that enable TESTING IMAGES to augument. Instead, we will be just add  a rescale attribute which is kind of standard while working with images that perform NORMALIZATION bringing all the pixels under same range [0,1] 

```python
## ImageGenerator object for TRAINING DATASET

train_datagen=ImageDataGenerator(
    shear_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True,
    rescale=1./255  #rescale performs normalization
)
```

```python
## Unlike earlier here we are using .flow_from_directory() here since we have a 

train_genrator=train_datagen.flow_from_directory(
    "/content/drive/MyDrive/cat-dog/training_set",  #representing the file where training dataset is present
    target_size=(150,150),
    batch_size=15,
    class_mode="binary"  #Since we have two category of images i.e Dog and Cat
)
```

```python
## As discussed earlier here we are only performing Normalization on top of Test dataset

test_datagen=ImageDataGenerator(
    rescale=1./255
)
```

```python
test_genrator=test_datagen.flow_from_directory(
    "/content/drive/MyDrive/cat-dog/test_set",
    target_size=(150,150),
    batch_size=15,
    class_mode="binary"
)
```

```python
## Creating new Sequential model and adding convoluted, polling flattening and neural network layers

model=Sequential()
```

```python
# this is my first convolution layer
model.add(Conv2D(24,kernel_size=(3,3),padding='valid',activation='relu',input_shape=(150,150,3)))
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

#second convolution
model.add(Conv2D(36,kernel_size=(3,3),padding='valid',activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

#third convlolution
model.add(Conv2D(48,kernel_size=(3,3),padding='valid',activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

#convert the data into 1-d array
model.add(Flatten())

#adding first hidden layer with 128 neurons and realu activation
model.add(Dense(64,activation='relu'))
#adding second hidden layer with 64 neurons and relu activation
model.add(Dense(32,activation='relu'))
#here we have a 1 output since it is a binary classification
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
## Here we are training our new CNN architecture using train_generator that consists of augumented images as well
## using .fit_generator 
## If you observe the warning that we are getting then you will understand that .fit would have also worked since it
## supports the generators

model.fit_generator(train_genrator,epochs=1,validation_data=test_genrator)
```
